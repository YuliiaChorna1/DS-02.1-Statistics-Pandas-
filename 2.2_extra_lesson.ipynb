### Step 2: Data Preparation

- dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns
- Feature Creation

```python
df.drop("Website", axis=1)
```

```python
df1 = df[['coaster_name', 
    # 'Length', 'Speed', 
    'Location', 'Status', 
    # 'Opening date', 'Type', 
    'Manufacturer', 
    # 'Height restriction', 'Model', 'Height',
    #    'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #    'Track layout', 'Fastrack available', 'Soft opening date.1',
    #    'Closing date', 'Opened', 'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
    'year_introduced', 'latitude', 'longitude', 'Type_Main',
    'opening_date_clean', 
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
    'speed_mph', 'height_value', 
    # 'height_unit', 'height_ft',
    'Inversions_clean', 'Gforce_clean']]
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
df1["opening_date_clean"] = pd.to_datetime(df1["opening_date_clean"])
```

```python
df1.dtypes
```

```python
df1.columns
```

```python
df1 = df1.rename(columns = {'coaster_name': 'Coaster_Name', 
                            'year_introduced': 'Year_Introduced',
                            'opening_date_clean': 'Opening_Date', 
                            'speed_mph': 'Speed',
                            'height_value': 'Height', 
                            'Inversions_clean': 'Inversions', 
                            'Gforce_clean': 'Gforce'})
```

```python
df1.head(3)
```

```python
df1.isna()
```

```python
df1.isna().sum()
```

```python
df1.loc[df1.duplicated(subset=["Coaster_Name"])].head()
```

```python
df1.query("Coaster_Name == 'Big Thunder Mountain Railroad'")
```

```python
df2 = df1.loc[~df1.duplicated(subset=["Coaster_Name", "Location", "Opening_Date"])]\
.reset_index(drop=True).copy()
```

```python
df2.shape
```

### Step 3: Feature Understanding

(Univariate analysis)

- Plotting Feature Distributions
    - Histogram
    - KDE
    - Boxplot

```python
df2["Year_Introduced"].value_counts().head(7)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,8))

ax = df2["Year_Introduced"].value_counts().head(10)\
    .plot(kind="bar", title="Top 10 Years Coaster Intro")

ax.set_xlabel("Years")
ax.set_ylabel("Number of Coasters")

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = df2["Speed"].plot(kind="hist", bins=20, title="Coaster Speed")
ax.set_xlabel("Speed")

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = df2["Speed"].plot(kind="kde", title="Coaster Speed")
ax.set_xlabel("Speed")

plt.show()
```

```python
df2["Type_Main"].value_counts()
```

### Step 4: Feature Relationships

- Scatterplot
- Heatmap Correlation
- Pairplot
- Groupby comparisons

```python
plt.figure(figsize=(12,8))
ax = df2.plot(kind="scatter", x="Speed", y="Height", title="Coaster Speed vs. Height")
ax.set_xlabel("Speed")

plt.show()
```

```python
df2.query("Height > 3500")
```

```python
df2 = df2.drop([836], axis=0)
```

```python
plt.figure(figsize=(12,8))
df2.plot(kind="scatter", x="Speed", y="Height", title="Coaster Speed vs. Height")
ax.set_xlabel("Speed")

plt.show()
```

```python
ax = sns.scatterplot(x="Speed", y="Height", hue="Type_Main", data=df2)
ax.set_title("Coaster Speed vs. Height")
plt.show()
```
