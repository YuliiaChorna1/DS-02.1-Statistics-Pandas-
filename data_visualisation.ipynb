### Візуалізація даних

Для візуалізації даних у бібліотеці pandas існує метод ***plot***, який викликається в об'єкта структури Series або DataFrame. 

Якщо використовувати цей метод без параметрів, то буде побудований **лінійний графік**.

За вигляд графіка відповідає аргумент ***kind*** і, залежно від його значення, змінюватиметься форма графічного представлення даних.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date = pd.date_range(start="2021-09-10", freq="D", periods=8)

temperature = pd.Series([23, 17, 17, 16, 15, 14, 17, 20], index=date)

temperature.plot()

plt.show()
```
